```python
def check_entity_dom(dom_ent_dict, dom_label, df):
    entities = dom_ent_dict[dom_label]
    df_filtered = df[df["entity_id"].isin(entities)]
    return df_filtered
    
check_entity_dom(dom_ent, dom_list[37], df_ent_info)

################
#   0   language
#   1
#   2
#   3   place
#   4   place
#   5   place
#   6   person
#   7   person
#   8   
#   9   film, actor
#   10  TV series, actor
#   11  
#   12  film, actor
#   13  university
#   14  
#   15  
#   16  film
#   17  
#   18  air port
#   19  singer, musician
#   20  space, defense organization
#   21  place
#   22  country, army, war
#   23  place
#   24  
#   25  actor
#   26  actor
#   27  singer, actor
#   28  football club, player
#   29  football club
#   30  actor, musician
#   31  actor
#   32  writer
#   33  country, place, football player
#   34  place
#   35  
#   36  church
#   37  country
```
