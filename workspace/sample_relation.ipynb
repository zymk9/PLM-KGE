```python
import json
import random
```

```python
f = open('data/dom_ent.json')
dom_ent = json.load(f)

f = open('data/ent_dom.json')
ent_dom = json.load(f)

f = open('data/rel2dom_h.json')
rel2dom_h = json.load(f)

f = open('data/rel2dom_t.json')
rel2dom_t = json.load(f)

f = open('data/rel2nn.json')
rel2nn = json.load(f)

```

```python
for relation in rel2nn:
    # As SimKGC reverse the example. Each relation will be sampled twice: head and tail
    # relation_complexity:
    #   0:  1-1
    #   1:  1-N
    #   2:  N-1 
    #   3:  N-N
    
    # sample head, if the selected 
    
    print(rel2nn[relation])
```

```python
t_concept_list = rel2dom_t["actor regular tv appearance regular cast tv program tv "]
ent_list = []
ent_concepts_list = []
for i in range(1000):
    t_concept = random.choice(t_concept_list)
# for t_concept in t_concept_list:
    ent = random.choice(dom_ent[str(t_concept)])
    ent_list.append(ent)
    ent_concepts_list.append(ent_dom[ent])
    
scores = []
cnt = 0
for ent_concept_1 in ent_concepts_list:
    for ent_concept_2 in ent_concepts_list:
        if ent_concept_1 == ent_concept_2:
            continue
        set_1 = set(ent_concept_1)
        set_2 = set(ent_concept_2)
        desired_set_2 = set(t_concept_list) - set(ent_concept_1)
        scores.append(
            len(desired_set_2.intersection(set_2))/
            len(set_2)
        )
        cnt += 1

        
print(sum(scores) / cnt)
print(cnt)        

```

```python
################### check relation-example distribution ##################
f = open('data/train.txt.json', encoding="utf-8")
train_exs = json.load(f)

rel_ex_distribution = {}
for ex in train_exs:
    if ex["relation"] in rel_ex_distribution:
        rel_ex_distribution[ex["relation"]] += 1
    else:
        rel_ex_distribution[ex["relation"]] = 1
        
print(rel_ex_distribution)
```

```python
relations = list(rel_ex_distribution.keys())
relation_ex_cnt = list(rel_ex_distribution.values())

sampled = random.choices(
    relations,
    weights = relation_ex_cnt,
    k = 1
)
print(rel_ex_distribution[sampled[0]])
```
